**Question:**
After 10 repetitions of a stretch, does one’s flexibility improve in a single session?


**How:**
First in a seated “touch your toes” stretch, measure the distance from fingertips to toes in inches and logged manually. Then stretcher will conduct 10 more repetitions of the stretch. After these stretch repetitions, the distance will be measured in inches.


If flexibility is improved in a single session, then the user will select button A.

Or if flexibility remained the same or got worse, then user will select button B.

User's selection will count "1" into either the "BETTER" or "SAME_OR_WORSE" column to indicate each result.

```python
# import libraries

import pandas as pd # library for reading and manipulating data
import matplotlib.pyplot as plt # library for visualization abbreviated
from matplotlib.pyplot import pie, show # pulling from matplotlib library the pie chart
import io # to enable ability to input / output specicially files in this case
from google.colab import files # library to upload files onto colab

# This code will create a button in the output section. Click it and select your file
CSVNAME = "FLEXBIT_DATA_LOG .csv" # this name changes to match the number of times this file was uploaded
# uploaded = files.upload()

import os

files = os.listdir()

flexbitData = pd.read_csv(files[1]) # assigning the results of the CSV file read into a variable
```

```python
# define global variables in this cell

#storing the sum of counts in each column ('BETTER' vs 'SAME_OR_WORSE')respectively
count_of_better = flexbitData.BETTER.sum()
count_of_same_or_worse = flexbitData.SAME_OR_WORSE.sum()

# assign the data to variables
label_name = ['Better', 'Same or Worse'] # clarifying label names
colors = ['#67B346', '#82D8D5'] # an aesthetically pleasing hue of green and blue
results = [count_of_better, count_of_same_or_worse] # results are the sum of the counts in BETTER vs. SAME_OR_WORSE comments

flexbitData.head(10) # ensuring information from the CSV has been retrieved
```

```python
# define classes and functions

# Build the pie chart
# This is where AI-generated code helped create an inital pie chart. Class structure was built through internet research.
# 4 total lines were AI-generated and every one of those line has been modified. Prompt used: "create a pie chart with the two above variables"

class pieChart:
  def __init__(self, data, labels):
    self.data = flexbitData
    self.labels = label_name
# this is the initializing method for the class pieChart which presents the data from flexbitData and assigns them with labels

  def chart(self):
    plt.figure(figsize=(9, 9)) # AI generated line with modifications. size the chart for readibility
    explode = (0, 0.07) # explode the "better" slice out from the pie
    plt.pie(results, labels = label_name, explode = explode, shadow = True, colors = colors, autopct = '%1.1f%%', startangle = 300) # AI generated line with modifications. Stylize the chart.
    plt.title('Improvements After Dynamic Stretching') # AI generated line with modifications. Good practice to label charts.
    plt.show() # AI generated line with modifications. call the method to show the chart
```

```python
# execute code in this cell

flexbitDataObject = pieChart(flexbitData, label_name) # creating an object off the pieChart class with Flexbit Data
flexbitDataObject.chart() # call the method to show the chart with Flexbit data
```
